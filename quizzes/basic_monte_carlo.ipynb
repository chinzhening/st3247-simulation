```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# q1
# int_(-1)^(1) int_(-1)^(1) exp(sin(x^2 + y^2)^2) dy dx

n = 10_000

x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)

f = np.exp(np.sin(x**2 + y**2)**2) * 4
I = np.mean(f) 

print("Estimated integral:", I)
```

```python
# q2
x = np.random.uniform(-1, 5, n)

f = np.exp(np.sin(3 * x)) * 6
I = np.mean(f)

print("Estimated integral:", I)
```

```python
# q3

# estimate the standard deviation of X exp(-U) where X ~ N(0,1) and U ~ U(0,1)
x = np.random.normal(0, 1, n)
u = np.random.uniform(0, 1, n)

f = x * np.exp(-u)
std = np.std(f)

print("Estimated standard deviation of the mean:", std)
```

```python
# q4

# estimate volume of region defined by x^2 + 2 y^2 + 3 z^2 <= 1
x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)
z = np.random.uniform(-1, 1, n)

# this defines a bounding volume

inside = (x**2 + 2 * y**2 + 3 * z**2) <= 1
volume = 8 * np.mean(inside)

print("Estimated volume of the region:", volume)
```

```python
# q5

x = np.random.uniform(0, 2, n)
f = np.exp(-x) * 2

Z = np.mean(f)
print("Estimated normalizing constant Z:", Z)
```

```python
# q6

x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)

f = (2 + np.sin(x**2 + y**2)) * 4
Z = np.mean(f)

print("Estimated normalizing constant Z:", Z)
```

```python
# q7

n = 1_000_000

x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)

z = np.abs(x - y)
prob = np.mean(z <= 0.1)

print("Estimated probability P(|X - Y| <= 0.1):", prob)
```

```python
# q8
m = 50

coin = np.random.choice([0, 1], size=(n, m))
heads = np.sum(coin, axis=1)
prob = np.mean(heads == 25)

print("Estimated probability of getting exactly 25 heads in 50 tosses:", prob)
```

```python
# q9

x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)

u = x + np.sin(y)
v = np.exp(np.sin(x + y))

corr = np.corrcoef(u, v)[0, 1]
print("Estimated correlation between U and V:", corr)
```

```python
# q10

m = 10
u = np.random.uniform(0, 1, (n, m))
u_sum = np.sum(u, axis=1)

prob = np.mean(u_sum < 4)
print("Estimated probability P(U1 + U2 + ... + U10 < 4):", prob)
```

```python
# q11

x = np.random.uniform(0, 60, n)
y = np.random.uniform(0, 60, n)

meet = np.abs(x - y) <= 10

prob = np.mean(meet)
print("Estimated probability that two people meet within 10 minutes:", prob)
```

```python
# q13

Ns = []
n = 10_000
for i in range(n):
    total = 0
    N = 0
    while total <= 1:
        u = np.random.uniform(0, 1)
        total += u
        N += 1

    Ns.append(N)

mean_N = np.mean(Ns)
print("Estimated expected value of N:", mean_N)
```

```python
# q14

var_N = np.var(Ns)
print("Estimated variance of N:", var_N)
```

```python
# q15

m = 20
# biased coin with p = 0.6
coin = np.random.choice([0, 1], size=(n, m), p=[0.4, 0.6])
heads = np.sum(coin, axis=1)

prob = np.mean(heads >= 15)
print("Estimated probability of getting at least 15 heads in 20 tosses of a biased coin:", prob)
```

```python
# q16

p = np.random.uniform(0, 1, (n, 3))
q = np.random.uniform(0, 1, (n, 3))

D = np.linalg.norm(p - q, axis=1)

mean_D = np.mean(D, axis=0)
print("Estimated expected value of |P - Q| for each dimension:", mean_D)
```

```python
# q17
from scipy.stats import norm

# estimate int_(-oo)^(00) exp(-x^4) dx
x = np.random.normal(0, 1, n)
f = np.exp(-x**4) / norm.pdf(x)
I = np.mean(f)

print("Estimated integral int_(-oo)^(oo) exp(-x^4) dx:", I)
```

```python
# q18

m = 3
# biased die with prob proportional to face value
dice_rolls = np.random.choice([1, 2, 3, 4, 5, 6], size=(n, m), p=[1/21, 2/21, 3/21, 4/21, 5/21, 6/21])
three_sixes_in_row = np.all(dice_rolls == 6, axis=1)

prob = np.mean(three_sixes_in_row)
print("Estimated probability of rolling three sixes in a row with biased die:", prob)
```

```python
# q19

permutations = np.array([np.random.permutation(5) for _ in range(n)])
fixed_points = np.sum(permutations == np.arange(5), axis=1)

prob = np.mean(fixed_points == 0)
print("Estimated probability of no fixed points in a random permutation of 5 elements:", prob)
```

```python
# q20

x = np.random.uniform(0, 2, n)
y = x**2

hist, bin_edges = np.histogram(y, bins=50, range=(0, 4), density=True)
plt.bar(bin_edges[:-1], hist, width=bin_edges[1]-bin_edges[0])
plt.xlabel('y')
plt.ylabel('Estimated Density')
plt.title('Estimated PDF of Y = X^2 where X ~ U(0,2)')

xs = np.linspace(0, 4, 100)
pdf_theoretical = 1 / (4 * np.sqrt(xs))
plt.plot(xs, pdf_theoretical, 'r-', label='Theoretical PDF')
plt.xlabel('y')
plt.ylabel('Theoretical Density')
plt.title('Theoretical PDF of Y = X^2 where X ~ U(0,2)')
plt.legend()
plt.show()
```
